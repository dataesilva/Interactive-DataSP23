```python
import numpy as np
import pandas as pd
from bokeh.models import HoverTool
from bokeh.plotting import figure, show, reset_output
from bokeh.io import output_notebook, output_file
```

```python
twitch = pd.read_csv('https://raw.githubusercontent.com/Hamidest/InteractiveData_EMAT/main/twitchdata-update.csv')
twitch.head()
```

```python
twitch_new = twitch.groupby('Language')
twitch_new.count()
```

```python
counts = twitch_new.count().sort_values('Channel', ascending = False)
counts.head()
```

```python
counts.index
```

```python
x = list(counts.index)
y = counts['Channel']
output_notebook()
output_file('in-class-figure.html')

p1 = figure(x_range = x, sizing_mode = 'stretch_width')
p1.line(x, y)

show(p1)
```

```python
hover1 = HoverTool(tooltips=[('Number of Streamers', '@y')],
                  mode = 'vline')
reset_output()
p2 = figure(x_range = x, sizing_mode = 'stretch_width', height = 300,
           title = 'Twitch Streamers by Spoken Language', x_axis_label = 'Spoken Language',
           y_axis_label = 'Number of Streamers', toolbar_location = None, tools = [hover1])
p2.line(x, y)

p2.ygrid.grid_line_alpha = 0.8
p2.ygrid.grid_line_dash = [6, 4]

show(p2)
```

```python
twitch_new.sum()['Stream time(minutes)'].sort_values(ascending = False)
```

```python
# Need to remap the x values to the new sorted values for watch time
y2 = twitch_new.sum().sort_values('Stream time(minutes)', ascending = False)['Stream time(minutes)']

hover2 = HoverTool(tooltips=[('Number of Streamers', '$y{0.00}')],
                  mode = 'vline')

p3 = figure(x_range = x, sizing_mode = 'stretch_width', height = 300,
           title = 'Twitch Streamers by Spoken Language', x_axis_label = 'Spoken Language',
           y_axis_label = 'Number of Streamers', toolbar_location = None, tools = [hover2])
p3.line(x, y2)

show(p3)
```
